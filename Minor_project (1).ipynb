**Standardization and Cross Validation**

```python
def st_vd_classifier():
    """
    Standardization + CV{5}
    """
    features = df.iloc[:,:-1]
    target = df.iloc[:,-1]
    target.replace(["R","M"],[0,1], inplace=True)
    target = np.asarray(target)

    # Feature Scaling
    sc = StandardScaler()
    standard_features = sc.fit_transform(features)

    models = [LogisticRegression(), DecisionTreeClassifier(), KNeighborsClassifier(), SVC(), \
              RandomForestClassifier(), xgb.XGBClassifier()]
    model_df = []

    for model in models:
        cross_validation_classifier = cross_val_score(model, standard_features, target, cv=5, scoring='accuracy')
        accuracy = round((cross_validation_classifier.mean())*100,2)
        model_df.append({"Model":model,"Accuracy": accuracy})

    model_df = pd.DataFrame(model_df,columns=["Model", "Accuracy"])

    return model_df
```

## Deep Neural Network

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from google.colab import drive
drive.mount('/content/drive')

data = pd.read_csv('/content/drive/MyDrive/sonar_data.csv', header=None)

y = data[60]
x = data.drop(columns=60, axis=1)

le = preprocessing.LabelEncoder()

y = le.fit_transform(y)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=(0.3))

from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout

model = Sequential()

model.add(Dense(units=32, activation='relu', input_dim=len(x_train.columns)))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')

model.summary()

model.fit(x_train, y_train, epochs=1000, batch_size=32, verbose=2)

y_hat = model.predict(x_test)

results = []
for valor in y_hat:
  if valor < 0.5:
    var = 0
  else:
    var = 1
  results.append(var)

results = np.array(results)

accuracy_score = accuracy_score (y_test, results)

print('Accuracy with Deep Learning:', accuracy_score)
```

# Gradient Boosting Results

```python
import xgboost as xgb

#splitting the data
from sklearn.model_selection import train_test_split

#Evaluation (Nothing is mentioned in Problem. So, I'm going to use accuracy score)
from sklearn.metrics import classification_report,accuracy_score

#standardization
from sklearn.preprocessing import StandardScaler

#Cross Validation
from sklearn.model_selection import cross_val_score

df=pd.read_csv('/content/drive/MyDrive/sonar_data.csv', header=None)
features = df.iloc[:,:-1]
target = df.iloc[:,-1]
target.replace(["R","M"],[0,1], inplace=True)
target = np.asarray(target)

# Feature Scaling
sc = StandardScaler()
standard_features = sc.fit_transform(features)

xg_X_train, xg_X_test, xg_y_train, xg_y_test = train_test_split(standard_features,target,test_size=0.3,
                                                        stratify=target, random_state=42)
classifier = xgb.XGBClassifier(objective='binary:logistic',booster='gbtree',
                               gamma=0.1,learning_rate=0.5,max_depth=1,reg_alpha=1.6,reg_lambda=3.2)
classifier.fit(xg_X_train,xg_y_train)
```

```python
xg_training_predict = classifier.predict(xg_X_train)
xg_test_predict = classifier.predict(xg_X_test)
```

```python
xg_training_score = round(accuracy_score(xg_training_predict,xg_y_train)*100,4)
xg_test_score = round(accuracy_score(xg_test_predict,xg_y_test)*100,4)
```

```python
print(f"Accuracy on training dataset: {xg_training_score}%")
print(f"Accuracy on testing dataset: {xg_test_score}%")
```

```python
#Classification Report
print(classification_report(xg_y_train,xg_training_predict))
```

```python
print(classification_report(xg_y_test,xg_test_predict))
```
